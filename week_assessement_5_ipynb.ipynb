```python
from google.colab import drive
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
drive.mount('/content/drive')
students = pd.read_csv('/content/drive/MyDrive/data/Students_Grading_Dataset .csv')
students.head(10)
```

```python
students.dtypes
```

```python
students.shape
```

```python
missing_values=(students.isnull().sum())
print(missing_values)
```

```python
missing_values_percentage = (students.isnull().sum() / len(students)) * 100
print(missing_values_percentage)
```

```python
students['Attendance (%)'] = students['Attendance (%)'].fillna(students['Attendance (%)'].mean())
students['Assignments_Avg'] = students['Assignments_Avg'].fillna(students['Assignments_Avg'].mean())
students['Parent_Education_Level'] = students['Parent_Education_Level'].fillna(students['Parent_Education_Level'].mode()[0])
print(students.Parent_Education_Level)
```

```python
print(students.dtypes)
```

```python
# Drop unnecessary columns (e.g., 'First_Name', 'Last_Name', 'Email')
students = students.drop(columns=['First_Name', 'Last_Name', 'Email'])

# Verify the remaining columns
print(students.columns)
```

```python
# 2. Descriptive Statistics & Understanding Data
# What are the summary statistics (mean, median, std, min, max) of Midterm_Score, Final_Score, and Total_Score?
# Find the age distribution of students in the dataset.
# Count the number of students in each Department.
# What is the average Study Hours per Week for students in different departments?

summary_stats = students[['Midterm_Score', 'Final_Score', 'Total_Score']].describe()
print(summary_stats)
```

```python
import matplotlib.pyplot as plt
plt.hist(students['Age'], bins=10, edgecolor='black')
plt.title('Age Distribution of Students')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
department_counts = students['Department'].value_counts()
print(department_counts)
```

```python
avg_study_hours = students.groupby('Department')['Study_Hours_per_Week'].mean()
print(avg_study_hours)
```

```python
# 3. Academic Performance Analysis
# Which factors correlate most with Total_Score?
# How does Attendance (%) impact Final Scores?
# (Use a scatter plot to visualize)
# Does participation in extracurricular activities affect academic performance?
# Find out if students who study more hours per week score better in exams.
# (Create a visualization to compare Study_Hours_per_Week and Total_Score)
# numerical_students = students.select_dtypes(include=np.number)
# Correlation matrix of numerical columns
# corr_matrix = students.corr()
# Correlation matrix of numerical columns
# Exclude non-numeric columns, like 'Student_ID', 'Gender', 'Department', etc.
numeric_students = students.select_dtypes(include=np.number)
corr_matrix = numeric_students.corr()

# Display correlation with Total_Score
corr_matrix['Total_Score'].sort_values(ascending=False)
```

```python
# Scatter plot to show the relationship between Attendance and Final Score
sns.scatterplot(x='Attendance (%)', y='Final_Score', data=students)
plt.title('Impact of Attendance on Final Scores')
plt.xlabel('Attendance (%)')
plt.ylabel('Final Score')
plt.show()
```

```python
# Compare Total Score based on extracurricular participation
sns.boxplot(x='Extracurricular_Activities', y='Total_Score', data=students)
plt.title('Impact of Extracurricular Activities on Total Score')
plt.xlabel('Extracurricular Activities')
plt.ylabel('Total Score')
plt.show()
```

```python
# Scatter plot to compare Study Hours and Total Score
sns.scatterplot(x='Study_Hours_per_Week', y='Total_Score', data=students)
plt.title('Study Hours per Week vs Total Score')
plt.xlabel('Study Hours per Week')
plt.ylabel('Total Score')
plt.show()
```

```python
# 4. Social & Economic Factors Analysis
# Is there a relationship between Parent Education Level and Student Performance?
# Do students with Internet Access at Home perform better academically?
# Compare the performance of students from different Family Income Levels.
# Analyze the effect of Sleep Hours per Night on academic performance.
# (Plot a histogram to understand the distribution)
# Boxplot to compare Total Score by Parent Education Level
sns.boxplot(x='Parent_Education_Level', y='Total_Score', data=students)
plt.title('Parent Education Level vs Total Score')
plt.xlabel('Parent Education Level')
plt.ylabel('Total Score')
plt.xticks(rotation=45)
plt.show()
```

```python
# Do students with Internet Access at Home perform better academically?
# Compare Total Score based on Internet Access
sns.boxplot(x='Internet_Access_at_Home', y='Total_Score', data=students)
plt.title('Impact of Internet Access on Academic Performance')
plt.xlabel('Internet Access at Home')
plt.ylabel('Total Score')
plt.show()
```

```python
# Boxplot to compare Total Score by Family Income Level
sns.boxplot(x='Family_Income_Level', y='Total_Score', data=students)
plt.title('Family Income Level vs Total Score')
plt.xlabel('Family Income Level')
plt.ylabel('Total Score')
plt.show()
```

```python
# Plot the distribution of Sleep Hours vs Total Score
sns.histplot(students['Sleep_Hours_per_Night'], kde=True)
plt.title('Distribution of Sleep Hours per Night')
plt.xlabel('Sleep Hours per Night')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot for Sleep Hours and Total Score
sns.scatterplot(x='Sleep_Hours_per_Night', y='Total_Score', data=students)
plt.title('Sleep Hours per Night vs Total Score')
plt.xlabel('Sleep Hours per Night')
plt.ylabel('Total Score')
plt.show()
```

```python
# 5. Stress & Health Impact
# Is there a correlation between Stress Level and Final Scores?
# Find out if students who sleep less than 5 hours per night perform worse than others.
# Identify students who have high stress levels (>7) but high academic performance.
# (What characteristics do they share?)
students[['Stress_Level (1-10)', 'Final_Score']].corr()
```

```python
# Filter students who sleep less than 5 hours
sleep_less_5 = students[students['Sleep_Hours_per_Night'] < 5]

# Compare the performance of students who sleep less than 5 hours
sleep_less_5['Total_Score'].mean(), students['Total_Score'].mean()
```

```python
# Filter students with high stress but high performance
high_stress_high_perf = students[(students['Stress_Level (1-10)'] > 7) & (students['Total_Score'] > students['Total_Score'].median())]

# Analyze their characteristics
high_stress_high_perf.describe()
```

```python
# 6. Visualization & Insights
# Create a boxplot to show the distribution of Total_Score across different departments.
# Use a heatmap to visualize the correlation matrix of numerical columns.
# Identify the top 5 students based on Total_Score.
# Show the gender distribution in the dataset.
# Boxplot to compare Total Score by Department
sns.boxplot(x='Department', y='Total_Score', data=students)
plt.title('Distribution of Total Score by Department')
plt.xlabel('Department')
plt.ylabel('Total Score')
plt.show()
```

```python
import seaborn as sns

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Columns')
plt.show()
```

```python
# Sort the students by Total Score and get the top 5
top_5_students = students[['Student_ID', 'Total_Score']].sort_values(by='Total_Score', ascending=False).head(5)
top_5_students
```

```python
# Plot the gender distribution
sns.countplot(x='Gender', data=students)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
